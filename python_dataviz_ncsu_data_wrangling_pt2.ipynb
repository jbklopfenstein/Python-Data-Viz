```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
# using this to show stats such as the correlation coefficient between 2 variables
import scipy.stats as stats
# need this to explicitly register converters for future version of panda
from pandas.plotting import register_matplotlib_converters
print("Setup Complete")
```

```python
# Path of the file to read.  Note to self: current path is the path of jupyter files(which is C:\Users\jklopfen)
medmal_filepath = "./Documents/AAAA-DATA-WRANGLING/2-Data-Viz-Homework/MedMalpractice.csv"

#explicitly register matplotlib converters - required in future versions of panda
register_matplotlib_converters()

# Read the file into a variable fifa_data
medmal_data = pd.read_csv(medmal_filepath)
print("Done")
```

```python
# See what variables/columns we're working with
medmal_data.columns
```

```python
# Set the input variable X to 'Amount' only
# and print the statistical summary for initial analysis

X = medmal_data.Amount
print(X.describe())
```

```python
# Print the first 5 rows of the data
medmal_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Seaborn Histogram showing amounts of lawsuits 
sns.distplot(X)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Display the same in a bar chart for 'Insurance' type
sns.countplot(x=medmal_data.Insurance)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Display the same in a bar chart for 'Gender'
sns.countplot(x=medmal_data.Gender)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Analyze the relationship between Gender and Amount
sns.boxplot(x=medmal_data.Gender, y=medmal_data.Amount)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Analyze the relationship between Gender and Amount
sns.boxplot(x=medmal_data.Private_Attorney, y=medmal_data.Amount)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

sns.boxplot(x=medmal_data.Gender, y=medmal_data.Amount,
            hue=medmal_data.Private_Attorney, palette=["m", "g"],)
sns.despine(offset=10, trim=True)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Analyze the relationship between Gender and Amount
sns.scatterplot(x=medmal_data.Age, y=medmal_data.Amount)
```

```python
# Tried a jointplot

# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Analyze the relationship between Gender and Amount
sns.jointplot(x=medmal_data.Age, y=medmal_data.Amount)
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,10))

# Analyze the relationship between Gender and Amount
j = sns.jointplot(x=medmal_data.Age, y=medmal_data.Amount)
j.annotate(stats.pearsonr)
```
